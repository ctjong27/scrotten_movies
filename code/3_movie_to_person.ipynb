```python
import pandas as pd
import time
import requests
import os
from tqdm import tqdm

def get_movie_data(tmdb_id, api_key):
    url = f'https://api.themoviedb.org/3/movie/{tmdb_id}?api_key={api_key}'
    response = requests.get(url)
    data = response.json()
    return data

def get_movie_credits(tmdb_id, api_key):
    url = f'https://api.themoviedb.org/3/movie/{tmdb_id}/credits?api_key={api_key}'
    response = requests.get(url)
    data = response.json()
    return data

# Get current working directory
current_working_dir = os.getcwd()

# Construct the full path to your file
input_file_path = os.path.join(current_working_dir, "data", "oscar_movies_with_tmdb_id.csv")
output_file_path = os.path.join(current_working_dir, "data", "movie_to_persons.csv")

# load data
df = pd.read_csv(input_file_path)

# replace with your actual API key
api_key = '4ea37a8158ca566deb7455166dc1edda'

# Initialize the dataframe with the same columns
df_credits = pd.DataFrame(columns=['movie_id', 'imdb_id', 'person_id', 'person_name', 'department', 'role', 'character_or_job', 'credit_id'])

for tmdb_id in tqdm(df['tmdb_id']):
    movie_data = get_movie_data(int(tmdb_id), api_key)
    imdb_id = movie_data['imdb_id']
    credits_data = get_movie_credits(int(tmdb_id), api_key)

    for cast in credits_data['cast']:
        role = 'Cast_Uncredited' if 'uncredited' in cast['character'].lower() else 'Cast'
        row = {
            'movie_id': int(tmdb_id),
            'imdb_id': imdb_id,
            'person_id': cast['id'],
            'person_name': cast['name'],
            'department': 'Acting',  # Department for actors/actresses can be set to 'Acting'
            'role': role,
            'character_or_job': cast['character'],
            'credit_id': cast['credit_id']
        }

        df_credits = pd.concat([df_credits, pd.DataFrame([row])], ignore_index=True)
        # df_credits = df_credits.append(row, ignore_index=True)

    for crew in credits_data['crew']:
        if crew['department'].lower() in ['directing', 'production', 'writing']:
            row = {
                'movie_id': int(tmdb_id),
                'imdb_id': imdb_id,
                'person_id': crew['id'],
                'person_name': crew['name'],
                'department': crew['department'],  # This will set the 'department' to the respective department of the crew member
                'role': crew['job'],
                'character_or_job': crew['job'],
                'credit_id': crew['credit_id']
            }

            # df_credits = df_credits.append(row, ignore_index=True)
            df_credits = pd.concat([df_credits, pd.DataFrame([row])], ignore_index=True)

# Save the updated credits data
df_credits.to_csv(output_file_path, index=False)

# time.sleep(0.3)  # delay to avoid hitting rate limit
```
