```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('../data/tv_export_data.csv')

# Remove rows where the gender is 'Undetermined'
df = df[df['cast_popularity'] >= 20]
df = df[df['profile_race'] != '_total_']
df = df[df['profile_gender'] != 'Undetermined']

# Create a new column combining race and gender
df['race_gender'] = df['profile_race'] + '_' + df['profile_gender']

# Calculate the proportions for each show
df_prop = df.groupby(['show_id', 'race_gender']).episode_count.sum().unstack().fillna(0)
df_prop = df_prop.div(df_prop.sum(axis=1), axis=0)

# Create a correlation matrix
corr = df_prop.corr()

# Create a larger figure for the correlation matrix
plt.figure(figsize=(15, 15))

# Visualize the correlation matrix with a heatmap
sns.heatmap(corr, cmap='coolwarm', annot=True, square=True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('../data/tv_export_data.csv')

# Remove rows where the gender is 'Undetermined'
df = df[df['profile_gender'] != 'Undetermined']

# Filter to include only entries where the job is 'Acting'
df = df[df['known_for_department'] == 'Acting']

# Create a new column combining race and gender
df['race_gender'] = df['profile_race'] + '_' + df['profile_gender']

# Calculate the percentage representation for each show
df_percentage = df.groupby(['show_id', 'race_gender']).episode_count.sum().unstack().fillna(0)
df_percentage = (df_percentage.div(df_percentage.sum(axis=1), axis=0)) * 100

# Create a correlation matrix
corr = df_percentage.corr()

# Create a larger figure for the correlation matrix
plt.figure(figsize=(15, 15))

# Visualize the correlation matrix with a heatmap
sns.heatmap(corr, cmap='coolwarm', annot=True, square=True)

# Show the plot
plt.show()
```
