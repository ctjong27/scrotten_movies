```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os

data_folder = '../../data' 

path = os.path.join(data_folder, 'tv_season_to_race_gender_count.csv')
```

# normalize the result based on total ratio of each ethnic background from each other

```python
# Load the data
df = pd.read_csv(path)

# Remove 'Undetermined' columns if needed
df = df.filter(regex='^(?!.*_Undetermined).*$', axis=1)

# Drop columns that are not related to demographics (e.g., 'show_id', 'season_number')
demographics = df.drop(columns=['show_id', 'season_number'])

# Calculate the total sum of all demographics across the dataset
total_demographics_sum = demographics.sum().sum()

# Normalize the counts for each ethnic background by dividing by the total sum
normalized_demographics = demographics.div(total_demographics_sum) * 100
```

```python

# Now you can proceed as before, but use the normalized_demographics DataFrame
for demographic in normalized_demographics.columns:
    
    # Group by the demographic and compute the mean for all other demographics, groupby allowing for group by count of demographic group
    average_counts = normalized_demographics.groupby(df[demographic]).mean()

    # Visualize the data
    plt.figure(figsize=(15, 10))
    sns.heatmap(average_counts, cmap='coolwarm', annot=True)
    plt.title(f'Normalized Average Percentages by Ethnic Background for Varying Counts of {demographic}')
    plt.show()
```

```python
# Load the data
df = pd.read_csv(path)

# Remove 'Undetermined' columns if needed
df = df.filter(regex='^(?!.*_Undetermined).*$', axis=1)

# Drop columns that are not related to demographics (e.g., 'show_id', 'season_number')
demographics = df.drop(columns=['show_id', 'season_number'])

# Define a function to filter the middle 50% for each column (demographic group)
def filter_middle_50(col):
    lower_bound = col.quantile(0.25)
    upper_bound = col.quantile(0.75)
    return col[(col >= lower_bound) & (col <= upper_bound)]

# Apply the filtering function to each demographic group
filtered_demographics = demographics.apply(filter_middle_50)

# Normalize the counts for each ethnic background by dividing by the sum for each group (column)
normalized_demographics = filtered_demographics.div(filtered_demographics.sum()) * 100

# Transpose the DataFrame to have demographic groups as rows
normalized_demographics = normalized_demographics.T

# Visualize the data
plt.figure(figsize=(15, 10))
sns.heatmap(normalized_demographics, cmap='coolwarm', annot=True)
plt.title('Normalized Average Percentages by Ethnic Background for Middle 50% Counts')
plt.show()
```

```python
# Load the data
df = pd.read_csv(path)

# Remove 'Undetermined' columns if needed
df = df.filter(regex='^(?!.*_Undetermined).*$', axis=1)

# Drop columns that are not related to demographics (e.g., 'show_id', 'season_number')
demographics = df.drop(columns=['show_id', 'season_number'])

# Function to filter the middle 50% for unique values in each column (demographic group)
def filter_middle_50(col):
    unique_counts = col.dropna().unique()
    lower_bound = np.percentile(unique_counts, 25)
    upper_bound = np.percentile(unique_counts, 75)
    return col[(col >= lower_bound) & (col <= upper_bound)]

# Apply the filtering function to each demographic group
filtered_demographics = demographics.apply(filter_middle_50)

# Normalize the counts for each ethnic background by dividing by the sum for each group (column)
normalized_demographics = filtered_demographics.div(filtered_demographics.sum()) * 100

# Transpose the DataFrame to have demographic groups as rows
normalized_demographics = normalized_demographics.T

# Visualize the data
plt.figure(figsize=(15, 10))
sns.heatmap(normalized_demographics, cmap='coolwarm', annot=True)
plt.title('Normalized Average Percentages by Ethnic Background for Middle 50% Counts')
plt.show()
```
