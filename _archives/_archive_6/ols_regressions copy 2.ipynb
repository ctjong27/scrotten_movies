```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Assuming your data is saved in a CSV file named "tv_data.csv"
df = pd.read_csv("../data/tv_export_data.csv")

# Convert race into dummy variables
race_dummies = pd.get_dummies(df['profile_race'], prefix='race')
df = pd.concat([df, race_dummies], axis=1)

# Convert gender into a binary variable, and cast to integer
df['profile_gender'] = df['profile_gender'].map({'Female': 1, 'Male': 0, 'Undetermined': np.nan}).astype('Int64')

```
