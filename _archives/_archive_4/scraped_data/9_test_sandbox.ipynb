```python
# What is the percent of individuals listed that have missing profile pictures?
# This is found in movie_to_persons.csv and person_data.csv
# the columns are movie_to_persons.department and person_data.profile_path

# How does this relate to their department?
# How does this relate to the timeframe of the released movies?
# For individuals that have their ethnic background determined by their names, how does it correlate to the ethnic background determined by their image?
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Get current working directory
current_working_dir = os.getcwd()

# Construct the full path to your files
movie_to_persons_path = os.path.join(current_working_dir, "data", "movie_to_persons.csv")
person_data_path = os.path.join(current_working_dir, "data", "person_data.csv")
movie_data_path = os.path.join(current_working_dir, "data", "movie_data.csv")

# Load data
movie_to_persons_df = pd.read_csv(movie_to_persons_path)
person_data_df = pd.read_csv(person_data_path)
movie_data_df = pd.read_csv(movie_data_path)

# Merge movie_to_persons_df and person_data_df
merged_df = pd.merge(movie_to_persons_df, person_data_df, left_on='person_id', right_on='id', how='left')

# Merge with movie_data_df to get the release_date
merged_df = pd.merge(merged_df, movie_data_df[['id', 'release_date']], left_on='movie_id', right_on='id', how='left')

# Calculate the percentage of individuals with missing profile pictures
missing_profile_pct = (merged_df['profile_path'].isnull().sum() / len(merged_df)) * 100
print(f'Percent of individuals with missing profile pictures: {missing_profile_pct}%')

# Analysis based on department
department_analysis = merged_df.groupby('department')['profile_path'].apply(lambda x: (x.isnull().sum() / len(x)) * 100)
print('\nPercent of missing profile pictures by department:')
print(department_analysis)

# Analysis based on movie release year
# Ensure that 'release_date' is in datetime format
merged_df['release_date'] = pd.to_datetime(merged_df['release_date'])
merged_df['release_year'] = merged_df['release_date'].dt.year
year_analysis = merged_df.groupby('release_year')['profile_path'].apply(lambda x: (x.isnull().sum() / len(x)) * 100)

# Remove rows with missing release_year
year_analysis = year_analysis[~year_analysis.index.isnull()]

# Visualize the trend over years
year_analysis.plot()
plt.title('Percent of missing profile pictures by movie release year')
plt.xlabel('Release year')
plt.ylabel('Percent missing profile pictures')
plt.show()

departments = merged_df['department'].unique()

for department in departments:
    department_df = merged_df[merged_df['department'] == department]
    
    # Ensure that 'release_date' is in datetime format
    department_df['release_date'] = pd.to_datetime(department_df['release_date'])
    department_df['release_year'] = department_df['release_date'].dt.year
    year_analysis = department_df.groupby('release_year')['profile_path'].apply(lambda x: (x.isnull().sum() / len(x)) * 100)
    
    # Remove rows with missing release_year
    year_analysis = year_analysis[~year_analysis.index.isnull()]

    # Visualize the trend over years for the current department
    year_analysis.plot()
    # plt.title(f'Percent of missing profile pictures by movie release year for {department}')
    plt.title(f'Percent of missing profile pic {department}')
    plt.xlabel('Release year')
    plt.ylabel('Percent missing profile pictures')
    plt.show()
```

```python
# Load data
person_name_race_df = pd.read_csv("./data/person_name_race.csv")
person_mdb_image_race_df = pd.read_csv("./data/person_tmdb_image_race.csv")

# Mapping of original race categories to updated categories
mapping = {
    'white': 'white',
    'middle eastern': 'white',
    'latino hispanic': 'hispanic',
    'asian': 'api',
    'black': 'black',
    'indian': 'api'
}

# Apply the mapping to the 'image_race' column
person_mdb_image_race_df['updated_image_race'] = person_mdb_image_race_df['image_race'].map(mapping)

# Remove rows with null 'updated_image_race'
person_mdb_image_race_df = person_mdb_image_race_df[~person_mdb_image_race_df['updated_image_race'].isnull()]

# Merge dataframes on person_id
merged_df = pd.merge(person_name_race_df, person_mdb_image_race_df, on='id', how='inner')

# Create a new column 'race_agreement' which is True if 'name_race' and 'updated_image_race' are the same, False otherwise
merged_df['race_agreement'] = merged_df['name_race'] == merged_df['updated_image_race']

# Print the percentage of times 'name_race' and 'image_race' agree
agreement_percentage = 100 * merged_df['race_agreement'].sum() / len(merged_df)
print(f'The race determined by the name and the image agree {agreement_percentage:.2f}% of the time.')

# Calculate and print the percentage agreement for each race
race_agreement = merged_df.groupby('name_race')['race_agreement'].mean() * 100
print(race_agreement)
```

```python
# Load data
person_name_race_df = pd.read_csv("./data/person_name_race.csv")
person_mdb_image_race_df = pd.read_csv("./data/person_tmdb_image_race.csv")

# Mapping of original race categories to updated categories
mapping = {
    'white': 'white',
    'middle eastern': 'white',
    'latino hispanic': 'hispanic',
    'asian': 'api',
    'black': 'black',
    'indian': 'api'
}

# Apply the mapping to the 'image_race' column
person_mdb_image_race_df['updated_image_race'] = person_mdb_image_race_df['image_race'].map(mapping)

# Remove rows with null 'updated_image_race'
person_mdb_image_race_df = person_mdb_image_race_df[~person_mdb_image_race_df['updated_image_race'].isnull()]

# Merge dataframes on person_id
merged_df = pd.merge(person_name_race_df, person_mdb_image_race_df, on='id', how='inner')

# Create a confusion matrix
confusion_matrix = pd.crosstab(merged_df['name_race'], merged_df['updated_image_race'], margins=True)

# Print the confusion matrix
print("Actual data:")
print(confusion_matrix)

# Create a normalized confusion matrix
normalized_confusion_matrix = confusion_matrix / confusion_matrix.loc['All', 'All'] * 100

# Print the normalized confusion matrix
print("\nPercentage data:")
print(normalized_confusion_matrix)


# Create a normalized confusion matrix where each column sums to 100%
column_normalized_confusion_matrix = confusion_matrix.div(confusion_matrix.sum(axis=0), axis=1) * 100 * 2

# Print the column normalized confusion matrix
print("\nColumn percentage data:")
print(column_normalized_confusion_matrix)
```

# Given gender vs intrepreted gender
in person_data.csv, there is a gender column, see how it correlates to person_gender.csv [column predicted_gender] for when the gender is 1 or 2

```python
# Loading person_gender.csv
person_gender_path = os.path.join(current_working_dir, "data", "person_gender.csv")
person_gender_df = pd.read_csv(person_gender_path)

# Merging it with person_data_df on person_id
merged_gender_df = pd.merge(person_data_df, person_gender_df, left_on='id', right_on='id', how='left')

# Cleaning and encoding gender data
# Assuming predicted_gender is categorical data: 'male', 'female', and other categories
gender_encoding = {'male': 2, 'female': 1}
merged_gender_df['predicted_gender'] = merged_gender_df['predicted_gender'].map(gender_encoding)

# Filtering out rows where gender is 1 or 2 and not null
filtered_df = merged_gender_df[merged_gender_df['gender'].notna() & merged_gender_df['gender'].isin([1, 2])]

# Correlation analysis
gender_correlation = filtered_df[['gender', 'predicted_gender']].corr()

print(gender_correlation)

import numpy as np
print('')

# For person_data.csv
undetermined_gender_person_data = person_data_df[~person_data_df['gender'].isin([1, 2])].shape[0]
print(f'Total number of undetermined gender in person_data.csv: {undetermined_gender_person_data}')

# For person_gender.csv
# Assuming that undetermined gender is represented as NaN, if not please replace np.nan with the appropriate value
undetermined_gender_person_gender = person_gender_df[person_gender_df['predicted_gender'] == np.nan].shape[0]
print(f'Total number of undetermined gender in person_gender.csv: {undetermined_gender_person_gender}')
```
